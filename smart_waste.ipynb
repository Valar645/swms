```python
import pandas as pd

# Sample Smart Waste Bin Data
data = {
    'Date': ['2025-06-20', '2025-06-21', '2025-06-22', '2025-06-23', '2025-06-24', '2025-06-25', '2025-06-26'],
    'Bin1_Level (%)': [20, 40, 55, 70, 85, 95, 100],
    'Bin2_Level (%)': [10, 25, 45, 60, 78, 90, 98],
}

df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt

# Plot line graph
plt.plot(df['Date'], df['Bin1_Level (%)'], marker='o', label='Bin 1')
plt.plot(df['Date'], df['Bin2_Level (%)'], marker='o', label='Bin 2')

plt.title('Waste Bin Fill Levels Over Time')
plt.xlabel('Date')
plt.ylabel('Fill Level (%)')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Bar Chart to Compare Bin Levels
plt.figure(figsize=(8,5))
x = df['Date']
bin1 = df['Bin1_Level (%)']
bin2 = df['Bin2_Level (%)']

bar_width = 0.35
index = range(len(x))

plt.bar(index, bin1, bar_width, label='Bin 1', color='green')
plt.bar([i + bar_width for i in index], bin2, bar_width, label='Bin 2', color='blue')

plt.xlabel('Date')
plt.ylabel('Fill Level (%)')
plt.title('Bin 1 vs Bin 2 - Daily Fill Levels')
plt.xticks([i + bar_width / 2 for i in index], x, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Logic to detect full bins
df['Bin1_Status'] = df['Bin1_Level (%)'].apply(lambda x: 'Full' if x >= 90 else 'Not Full')
df['Bin2_Status'] = df['Bin2_Level (%)'].apply(lambda x: 'Full' if x >= 90 else 'Not Full')

df[['Date', 'Bin1_Level (%)', 'Bin1_Status', 'Bin2_Level (%)', 'Bin2_Status']]
```
